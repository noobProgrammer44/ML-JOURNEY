# CONCATENATION

```python
df1 =pd.DataFrame({
    "A": [1,2,3,4],
    "B": [10,20,30,40],
    "C": [100,200,300,400]
})
```

```python
df2 =pd.DataFrame({
    "A": [5,6,7,8],
    "B": [50,60,70,80],
    "C": [500,600,700,800]
})
```

```python
df3 =pd.DataFrame({
    "A": [9,10,11,12],
    "B": [90,100,110,120],
    "C": [900,1000,1100,1200],
    'D':[100,200,300,400]
})
```

```python
df=pd.concat([df1,df2,df3])
```

```python
df
```

MERGING


```python
df_try =pd.DataFrame({
    "A": [1,2,3,4],
    "B": [10,20,30,40],
    "C": [100,200,300,400]
})

df2_try =pd.DataFrame({
    "C": [100,200,300,400],
    "D": [50,60,70,80],
    "E": [500,600,700,800]
})
```

```python
pd.merge(df_try,df2_try)
```

# **FUNCTIONS**

```python
DF=pd.merge(df_try,df2_try)
```

```python
def multiplybyThree(x):
  return x*3
```

```python
multiplybyThree(DF)
```

```python
multiplybyThree(DF['A'])
```

```python
DF.apply(multiplybyThree)
```
